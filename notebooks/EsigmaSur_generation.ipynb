```python
import os
import matplotlib.pyplot as plt
from esigmapy.esigmaSur import EccentricSurrogate
```

```python
sur_data_dir = "/mnt/c/AKASH/git-repositories/github/esigmapy/esigmapy/esigmaSur/data" # Point to the directory where surrogate data is stored (esigmapy/esigmaSur/data)
ecc_sur_data_dir = os.path.join(sur_data_dir, "ecc_sur_data")
circ_sur_data_dir = os.path.join(sur_data_dir, "circ_sur_data")

sur = EccentricSurrogate(ecc_data_dir=ecc_sur_data_dir,
                        circ_data_dir=circ_sur_data_dir,
                        # verbose=True
                        )
```

```python
t, mode = sur( M = 10.,
     params_new = [5.3, 0.41, 3.], # q, e, l at t=t_ref
     delta_t = 1/2**12, 
     t_start=-136, 
     t_end=None, 
     remove_start_phase=True)

plt.figure()
plt.plot(t, mode.real)
plt.xlabel("Time (s)")
plt.ylabel(r"$\Re [h_{22}]$")
plt.show()
```
